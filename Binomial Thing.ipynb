```python
def u(k,i): #Working Day probability
    if k in [6, 7, 13, 14]:
        return 0
    else:
        return flatP(t= k-i)
            
def flatP(t, p0 = 0.318, changeDay = 3):
    if t <= changeDay:
        return p0
    else:
        return (14-t)*p0/(14-changeDay)
    
def uFunction(k,i): #Working Day probability
    if k in [6, 7, 13, 14]:
        return 0
    else:
        return flatP(t= k-i)

def failureProd(k,i):
    product = 1
    for l in range(i+1, k):
        product = product*(1-uFunction(l, i))
    return product

def constant(r, i):
    total = 0
    for k in range(r, i+8):
        total += uFunction(k,i)*failureProd(k,i)
    return total
```

```python
n = 10
pdf = []
cdf = []
i = 2
for k in range(n+1):
    thistotal = 0
    for j1 in range(k+1):
        for j2 in range(k-j1+1):
            for j3 in range(k-j1-j2+1):
                for j4 in range(k-j1-j2-j3+1):
                    for j5 in range(k-j1-j2-j3-j4+1):
                        for j6 in range(k-j1-j2-j3-j4-j5+1):
                                jtotal = j1 + j2 + j3 + j4 + j5 + j6
                                coeff = math.factorial(n)/(math.factorial(j1)*math.factorial(j2)*math.factorial(j3)*math.factorial(j4)*math.factorial(j5)*math.factorial(j6)*math.factorial(k-jtotal)*math.factorial(n-k))
                                temp = coeff*pow(u(i+1,i), j1)*pow((1-u(i+1,i)), (n-j1))*pow(u(i+2,i), j2)*pow((1-u(i+2,i)), (n-j1-j2))*pow(u(i+3,i), j3)*pow((1-u(i+3,i)), (n-j1-j2-j3))*pow(u(i+4,i), j4)*pow((1-u(i+4,i)), (n-j1-j2-j3-j4))*pow(u(i+5,i), j5)*pow((1-u(i+5,i)), (n-j1-j2-j3-j4-j5))*pow(u(i+6,i), j6)*pow((1-u(i+6,i)), (n-j1-j2-j3-j4-j5-j6))*pow(u(i+7,i), k-jtotal)*pow((1-u(i+7,i)), n-k)
                                thistotal += temp
    pdf.append(thistotal)
total = 0
for p in pdf:
    total += p
    cdf.append(total)
```
